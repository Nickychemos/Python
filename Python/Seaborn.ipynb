```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic')
df.head()
```

### Styling

```python
sns.set_style('darkgrid') #Making the grid dark. 'whitegrid can also be used'
sns.barplot(df, x = 'class', y = 'fare')
plt.show()
```

### Customization

```python
#Removing spines

sns.set_style('darkgrid')
sns.barplot(df, x = 'class', y = 'fare')
sns.despine()
plt.show()
```

```python
#Customizing theme parameters

sns.set_context('paper', font_scale = 1.4) #takk/text can be used
sns.barplot(df, x = 'class', y = 'fare')
plt.show()
```

```python
#color palletes

sns.barplot(df, x = 'class', y = 'fare')
sns.set_palette('coolwarm') #colorblind
plt.show()
```

```python
sns.palplot(sns.color_palette('coolwarm', 6))
```

```python
sns.palplot(sns.color_palette('coolwarm', 6))
```

```python
#Custom color palette

custom_palette =  sns.color_palette(['#FF5733', '#33FFCE', '#33FBFF'])
sns.set_palette(custom_palette)
sns.barplot(df, x = 'class', y = 'fare')
plt.show()
```

### Plotting

#### 1. Univariate plots

```python
#Histogram

df['age'].dropna()

sns.histplot(df, x = 'age')
```

```python
df['age'].dropna()

sns.displot(df, x = 'age')
```

```python
#kde

df['age'].dropna()

sns.displot(df, x = 'age', kind = 'kde')
plt.show()
```

```python
#kde + histplot

df['age'].dropna()

sns.displot(df, x = 'age', kde = True)
```

```python
#kde using .kdeplot

df['age'].dropna()

sns.kdeplot(df, x = 'age', fill = True)
```

```python
#Rugplot
df['age'].dropna()

plt.figure(figsize = (14,8))
sns.rugplot(df, x = 'age')
plt.show()
```

### 2. Bivariate

```python
#Jointplot

plt.figure(figsize = (12,20))
sns.jointplot(df, x = 'age', y = 'fare', kind = 'scatter')
plt.show()
```

```python
#Pairplots

#1.
df['age'].dropna()
sns.pairplot(df[['age', 'fare', 'class']], hue = 'class') # Pass numeric variables
plt.show()
```

```python
#2.
sns.pairplot(df[['age', 'fare', 'parch']]) # Pass numeric variables
plt.show()
```

### Visualizing Categorical Data

```python
# 1. Barplot
sns.barplot(df, x = 'class', y = 'fare')
plt.show()
```

```python
import numpy as np

sns.barplot(df, x = 'class', y = 'fare', estimator = np.median, ci = 'sd', palette = 'muted')
plt.show()
```

```python
#C2. Countplot --->vertical 

sns.countplot( df, x = 'class')
plt.show()
```

```python
#Countplot --->horizontal 

sns.countplot( df, y = 'class')
plt.show()
```

```python
#3. Boxplot

sns.boxplot(df, x = 'class', y = 'age')
```

```python
sns.boxplot(df, x = 'class', y = 'age', hue = 'sex')
plt.show()
```

```python
#4. Violinplots ---> combines kde and boxplot

sns.violinplot(df, x = 'class', y = 'age', hue = 'sex')
plt.show()
```

```python
# 5. Swarmplot -->Avoids overlap

sns.swarmplot(df, x = 'class', y = 'age')
plt.show()
```

```python
# 6. stripplot

sns.stripplot(df, x = 'class', y = 'age', jitter = True)
plt.show()
```

```python
sns.stripplot(df, x = 'class', y = 'age')
plt.show()
```

```python
#7. Scatter
sns.scatterplot(df, x = 'age', y = 'fare', hue = 'sex', size = 'pclass', style = 'embarked')
plt.show()
```

```python
#8. lineplot

sns.lineplot(df, x = 'age', y = 'fare', hue = 'sex')
plt.show()
```

```python
# 9. Regplots (regression plots)

sns.regplot(df, x = 'age', y ='fare')
plt.show()
```

```python
#10. Heatmap


data = df.pivot_table(values = 'fare', aggfunc = 'mean', index = 'class', columns = 'sex')

sns.heatmap(data, cmap = 'coolwarm', annot = True)
plt.show()
```

```python
#Using correlation matrix

corr_matrix = df.corr(numeric_only = True) #Shortcut for getting numeric values

corr_matrix

plt.figure(figsize = (11,7))
sns.heatmap(corr_matrix, cmap = 'coolwarm', annot = True, fmt = '.2f')
plt.title('A Heatmap of Titanic Dataset')
plt.show()
```
