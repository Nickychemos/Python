```python
#Determine the percentage of orders that resulted in a loss (negative profit).

df['profit_and_losses'] = df['Profit']<0       #creating a new column to show True and False for losses

number_of_losses = df.groupby('profit_and_losses').get_group(True).count()        #getting the total number of Trues
total_profit_and_losses = df['profit_and_losses'].count()          #getting the total number of profits and losses

percentage_of_losses = number_of_losses/total_profit_and_losses*100         #finding the percentage of trues among the total number of profit and losses
percentage_of_losses      # Losses were 18.72%
```
